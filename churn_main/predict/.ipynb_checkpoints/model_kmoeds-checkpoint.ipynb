### Kmodes Clustering 

```python
X_train = new_data

```

```python
# Elbow curve to find optimal K
cost = []
K = range(1,5)
for num_clusters in list(K):
    kmode = KModes(n_clusters=num_clusters, init = "random", n_init = 5, verbose=1)
    kmode.fit_predict(X_train)
    cost.append(kmode.cost_)
    
plt.plot(K, cost, 'bx-')
plt.xlabel('No. of clusters')
plt.ylabel('Cost')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# Building the model with 3 clusters
kmode = KModes(n_clusters=3, init = "random", n_init = 5, verbose=1)
clusters = kmode.fit_predict(X_train)
clusters
```

```python
raw_data.insert(0, "Cluster", clusters, True)
raw_data
```

```python
# calculate the churn rate for each cluster
churn_rate_cluster0 = raw_data.loc[raw_data['Cluster']== 0]
churn_rate0 = len(churn_rate_cluster0.loc[churn_rate_cluster0['Attrition_Flag']== 'Attrited Customer'])*100/len(raw_data)
print(churn_rate0)

churn_rate_cluster1 = raw_data.loc[raw_data['Cluster']== 1]
churn_rate1 = len(churn_rate_cluster1.loc[churn_rate_cluster1['Attrition_Flag']== 'Attrited Customer'])*100/len(raw_data)
print(churn_rate1)

churn_rate_cluster2 = raw_data.loc[raw_data['Cluster']== 2]
churn_rate2 = len(churn_rate_cluster2.loc[churn_rate_cluster2['Attrition_Flag']== 'Attrited Customer'])*100/len(raw_data)
print(churn_rate2)
```
